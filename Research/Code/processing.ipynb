```python
import pandas as pd

data = pd.read_csv('sp500_transcripts_history.csv')
data.head()
```

```python
data.info()
```

```python
data_2 = pd.read_csv('sp500_transcripts_history_part_2.csv')
data_2.info()
```

```python
# fusionner les deux datasets
data_full = pd.concat([data, data_2], ignore_index=True)
data_full.info()
```

```python
# afficher le nombre de 2025 dans la colonne 'year'
print((data_full['year'] == 2025).sum(),"number of 2025 earnings calls")
print((data_full['year'] == 2024).sum(),"number of 2024 earnings calls")
print((data_full['year'] == 2023).sum(),"number of 2023 earnings calls")
```

```python
print((data_full['quarter'] == 'Q1').sum(),"number of Q1 earnings calls")
print((data_full['quarter'] == 'Q2').sum(),"number of Q2 earnings calls")
print((data_full['quarter'] == 'Q3').sum(),"number of Q3 earnings calls")
print((data_full['quarter'] == 'Q4').sum(),"number of Q4 earnings calls")
```

```python
# afficher la liste des entreprises uniques
print(data_full['company'].nunique(),"number of unique companies")
print(data_full['company'].unique(),"list of unique companies")
```
